### Requirement 10

```python
# Use RandomState() and plot() to demonstrate 11 different plot marker 
# types. Also, display the markers in a legend.

%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
```

```python
rng = np.random.RandomState()
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(11), rng.random(11), marker,
            label="Pmarker'{0}'".format(marker))
plt.legend(numpoints=1)
plt.xlim(0, 1.9);
```

### Requirement 15

```python
#  Produce the following plot (purple and orange).
def f(x, y):
    return np.sin(x)** 10 + np.cos(10 + y *x) * np.cos(x)
```

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 40)

x, y = np.meshgrid(x, y)
z =f(x, y)
```

```python
contours = plt.contour(x, y, z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)

plt.imshow(z, extent=[0, 5,0, 5], origin='lower',
          cmap='PuOr', alpha=0.5)
plt.colorbar();
```

### Requirement 17

```python
# Produce the following plot (notice the color and legends).
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
import matplotlib as Legend
x = np.linspace(0, 10, 1000)
fig, ax = plt.subplots()

lines = []
styles = ['-', '--', '-.', ':']
x = np.linspace(0, 10, 100)

for i in range(4):
    lines += ax.plot(x, np.sin(x - i * np.pi / 2),
                     styles[i], color='#87C385')

ax.axis('equal')

ax.legend(lines[:2], ['line A', 'line B'],
         loc='upper left', frameon=False)

# creating second ledgen
from matplotlib.legend import Legend
leg = Legend(ax, lines[2:], ['line C', 'line D'],
             loc='lower left', frameon=False)
ax.add_artist(leg);
```
